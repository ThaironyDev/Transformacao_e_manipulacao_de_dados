##Projeto desafio 1: vendas online


###Normalizando dados

```python
dados_vendas = pd.read_json(url_1)
dados_vendas.columns
```

```python
dados_vendas.head()
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas.head()
```

###Remover os dados em listas dentro do DataFrame;

```python
selecao = list(dados_vendas.columns[1:])
selecao
```

```python
dados_vendas = dados_vendas.explode(selecao)
```

```python
dados_vendas.head()
```

###Verificar os tipos de dados

```python
dados_vendas.info()
```

###Identificar colunas numéricas

```python
dados_vendas.head()
```

###Transformar a coluna numérica para o tipo numérico

```python
import numpy as np
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip()).apply(pd.to_numeric, errors='coerce').astype(np.float64)
```

```python
dados_vendas.head()
```

###Redefinindo o indice

```python
dados_vendas.reset_index(inplace=True, drop=True)
```

```python
dados_vendas.head()
```

###Devido a uma instabilidade no site da empresa, tivemos problemas com os nomes dos clientes durante o salvamento. Isso resultou em uma coluna de nomes de clientes com uma mistura de letras, maiúsculas e minúsculas, números e outros caracteres.
Sabendo disso, manipule os textos presentes na coluna Cliente para que seja obtido como resultado os nomes dos clientes em letras minúsculas, com a ausência de caracteres especiais ou números.

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower().str.replace(r'[^a-z ]', '', regex=True).str.strip()
#Neste trecho estamos removendo caracteres especiais e numericos assim como espaços em branco no final de cada valor.
```

```python
dados_vendas.tail()
```

###Na coluna Data de venda, temos datas em formato 'dia/mês/ano' (dd/mm/AAAA). Transforme esses dados para o tipo datetime e busque uma forma de visualização de subconjunto que possa contribuir no objetivo do contexto que os dados estão inseridos.

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'])
dados_vendas.info()
```

```python
selecao = dados_vendas['Data de venda'].dt.strftime('%Y/%m')
dados_vendas.groupby(selecao)[['Cliente', 'Valor da compra']].max().groupby('Cliente')[['Valor da compra']].max().sort_values('Valor da compra', ascending=False)
```

##Projeto desafio 2: administração de condomínios

###Normalizando dados

```python
dados = pd.read_json(url_2)
dados.columns
```

```python
dados = pd.json_normalize(dados['dados_locacao'])
dados.head()
```

###Remover os dados em listas dentro do DataFrame

```python
dados.head()
```

```python
selecao = list(dados.columns[1:])
selecao
```

```python
dados = dados.explode(selecao)
```

```python
dados.head()
```

###Verificar os tipos de dados

```python
dados.info()
```

###Identificar colunas numéricas

```python
dados.head()
```

###Transformar a coluna numérica para o tipo numérico

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip()).apply(pd.to_numeric, errors='coerce').astype(np.float64)
dados.head()
```

###Redefinindo o índice

```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

###Buscando explicar a organização da identificação dos apartamentos, durante a criação do conjunto de dados, foi adicionado o texto (blocoAP). Esse texto informa que os nomes dos apartamentos estão organizados com a letra do bloco seguida do número do apartamento. No entanto, isso não traz nenhuma informação para nossos dados, sendo interessante realizar a remoção desse texto no conjunto de dados.

Com isso, manipule os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.

```python
dados['apartamento'] = dados['apartamento'].str.replace(r'\(blocoAP\)', '').str.strip()
#Neste trecho estamos removendo um padrão de string usando o método replace com uma regex e finalizando a limpeza com o métod strip().
```

```python
dados.head()
```
