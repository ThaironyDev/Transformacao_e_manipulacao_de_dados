```python
import pandas as pd
import requests
```

```python
url_1 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json'
url_2 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json'
```

##Normalizando dados de acordo com a coluna info_moveis.

```python
dados_hospedagem = pd.read_json(url_1)
dados_hospedagem.columns
```

```python
dados = pd.json_normalize(dados_hospedagem['info_moveis'])
dados.head()
```

##Desagrupando dados

>Podemos notar que algumas colunas possuem valores contidos em listas, precisamos desagrupar esses valores e para isso podemos usar o método explode(), esse método pertence ao objeto DataFrame e ele espera receber os nomes das colunas que contêm os valores agrupados, o retorno desse método é um novo dataframe ao com novos registros que são os valores que antes estavam contidos nas listas formando novas linhas.

```python
colunas = list(dados.columns[3:])
colunas
```

```python
dados = dados.explode(colunas)
dados.head()
```

>Resetando os índices para ordenar os dados.


---
O parâmetro drop=True no método reset_index() quer dizer que a coluna que originalmente continha os índices antigos será removida do DataFrame, e os índices serão redefinidos para os padrões (0, 1, 2, ...)



```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

>Verificando os tipos de dados em cada coluna do dataframe para entender qual tipo de limpeza será necessária.

```python
dados.info()
```

##Convertendo tipo

```python
import numpy as np
```

>Tratando dados numéricos: convertendo para o tipo inteiro.

```python
colunas_a_converter = ['max_hospedes', 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[colunas_a_converter] = dados[colunas_a_converter].apply(lambda x: x.astype(np.int64))
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

```python
dados['avaliacao_geral']
```

```python
selecao = ['taxa_deposito', 'taxa_limpeza', 'preco']
'''
O método applymap() vai ser aplicado para cada valor desse dataframe() esse método é o mais indicado pois não se trata de uma seleção de apenas uma Serie.
Depois do tratamento de cada valor do dataframe utilizando o método apply temos o tratamento das colunas, no apply() temos o método pd.to_numeric sendo passado
como uma função de callback para garantir que o conteúdo de cada coluna tenha valores numéricos e caso não tenham o parâmetro errors="coerce" para um NaN seja
inserido nesse lugar. Por fim o tipo é alterado  como astype().
'''
dados[selecao] = dados[selecao].applymap(lambda x: x.replace('$', '').replace(',', '').strip()).apply(pd.to_numeric, errors='coerce').astype(np.float64)
```

```python
dados.info()
```

```python
dados[selecao]
```

##Análise textual

>Tokenização

---
A tokenização refere-se ao processo de dividir um texto ou uma sequência de caracteres em unidades menores, conhecidas como "tokens". Estes tokens podem ser palavras, frases, sílabas ou outros elementos, dependendo do contexto. O objetivo da tokenização é facilitar a análise e manipulação de texto, tornando mais fácil para os computadores entenderem e processarem a linguagem humana. Em processamento de linguagem natural, por exemplo, a tokenização é frequentemente usada como etapa inicial para análise de texto, permitindo que algoritmos e modelos compreendam melhor a estrutura e o significado do conteúdo textual.


```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
#Neste trecho estamos usando o metodo lower() para converter os caracteres maiúsculos em minusculos, esse método pertence ao objeto String.
```

```python
dados.head()
```

>Substituindo caracteres usando a Regex e o método replace().


```python
dados['descricao_local'][3169]
#Neste trecho estamos visualizado um índice aleatório para visualizar os caracteres e enterder quais devem ser removidos.
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_local']
#Neste trecho estamos substituindo qualquer caracter que não correponda aos padrões informados na regex.
```

>Removendo ifens que não estejam acompanhando palavras compostas.

```python
dados['descricao_local'] = dados['descricao_local'].str.replace(r'(?<!\S)-(?!\S)', ' ', regex=True)
dados['descricao_local']
#Neste trecho estamos substituindo o hífen que não faça parte de uma palavra composta.
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
#Neste trecho para cada valor da coluna selecionada estamos traformando seu valor em uma lista de valores de string usando o método split()
```

>Tratando a coluna comodidades.

```python
dados['comodidades'] = dados['comodidades'].str.replace(r'[{}"]', '', regex=True).str.split(',')
#Neste trecho estamos substituindo caracteres indesejados usando o método str.replace() e a regex para indenficar esses padrões, por fim traformamos cada valor em uma lista de palavras.
```

```python
dados[['comodidades']]
```

>Tratando a coluna descricao_vizinhanca


```python
dados['descricao_vizinhanca'][0]
#Neste trecho estamos analisando a descrição em busca de caracteres indesejados.
```

```python
dados['descricao_vizinhanca'].str.contains(r'(?<!\S)-(?!\S)', regex=True).sum()
#Neste trecho estamos verificando se existem hífen isolados em algum valor da coluna selecionada.
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(r'[^a-zA-Z0-9\'\-]|(?<!\S)-(?!\S)', ' ', regex=True)
#Neste trecho estamos realizando o tratamento de caracteres indesejados substituindo por um espaço em branco mas preservando os hífens em palavras compostas.
```

```python
dados['descricao_vizinhanca']
```

#Trabalhando com o tipo DateTime.

```python
dt_datas = pd.read_json(url_2)
dt_datas.head()
```

```python
dt_datas.info()
```

>Usando o método pd.to_datetime() para converter o tipo da coluna data de object para datetime.

```python
dt_datas['data'] = pd.to_datetime(dt_datas['data'])
```

```python
dt_datas.info()
```

```python
dt_datas.head()
```

```python
selecao = dt_datas['data'].dt.strftime('%Y-%m')
selecao
#Nestre trecho usando o método strftime() estamos manipulando o valor da coluna data para visulizar apenas o ano e o mês de acordo com o formato especificado.
```

>De acordo com a seleção feita anteriormento e instanciada na variável selecao, vamos agrupar com base na seleção de mêses do ano os valores da coluna 'vaga_disponivel' esses valores são booleanos de maneira que ao utilizarmos o método sum() podemos somar os valores True e perceber o numero de vagas disponíveis para que aquele mês de um dados ano.

```python
subset = dt_datas.groupby(selecao)['vaga_disponivel'].sum()
subset
#Neste trecho estamos criando uma variável subset.
```

###Transformando os dados da coluna preco para o formato float64.

```python
dt_datas.info()
```

```python
dt_datas.tail()
```

```python
'''Neste trecho utilizando o metodo replace() com uma regex identificamos tudo que não éra um numero ou . para remover, garantindo o resultado usamos os método
strip() para remover spaços em branco das laterais, com o fillna() substituimos valores nulos por 0.0 e por fim alteramos o tipo usando o método astype()'''
import numpy as np
dt_datas['preco'] = dt_datas['preco'].str.replace(r'[^0-9.]', '', regex=True).str.strip().fillna(0.0).astype(np.float64)
```
